# Assignment 3: English to bengali translation

Write a Python program using the Hugging Face Transformers library to load the pretrained machine translation model Helsinki-NLP/opus-mt-en-bn and perform inference for English-to-Bengali translation.

Your code should load the tokenizer and model, take an English sentence as input, generate the Bengali translation, and print both the input and translated output

```python
from transformers import MarianMTModel, MarianTokenizer

# Define model name
model_name = "shhossain/opus-mt-en-to-bn"

# Load tokenizer and model
tokenizer = MarianTokenizer.from_pretrained(model_name)
model = MarianMTModel.from_pretrained(model_name)

# Input English text
src_text = input("Enter English text: ")

# Preprocess and translate
translated = model.generate(**tokenizer(src_text, return_tensors="pt", padding=True))

# Decode and print results
for t in translated:
    print(tokenizer.decode(t, skip_special_tokens=True))
```
